```python
import pandas as pd

# CSV dosyasından çalışan verilerini yükleme
df = pd.read_csv('/content/drive/MyDrive/Dersler_2026/Veri/calisanlar.csv')

# İlk 5 satırı görüntüleme
print(df.head())

# Veri seti hakkında genel bilgi
print(df.info())
print(df.describe())
```

```python
# Eksik verileri kontrol etme
print(df.isnull().sum())
```

```python
# Eksik verileri doldurma
df['maas'].fillna(df['maas'].median(), inplace=True)
```

```python
# Veri tiplerini düzeltme
df['ise_giris_tarihi'] = pd.to_datetime(df['ise_giris_tarihi'])
```

```python
# Sütun grafiği - Departmanlara göre çalışan dağılımı
departmanlar = df['departman'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(departmanlar.index, departmanlar.values, color='steelblue')
plt.title('Departmanlara Göre Çalışan Dağılımı', fontsize=14)
plt.xlabel('Departman', fontsize=12)
plt.ylabel('Çalışan Sayısı', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram - Maaş dağılımı
plt.figure(figsize=(10, 6))
plt.hist(df['maas'], bins=20, color='coral', edgecolor='black', alpha=0.7)
plt.title('Maaş Dağılımı', fontsize=14)
plt.xlabel('Maaş (TL)', fontsize=12)
plt.ylabel('Çalışan Sayısı', fontsize=12)
plt.axvline(df['maas'].median(), color='red', linestyle='--',
            label=f'median: {df["maas"].mean():.0f}')
plt.legend()
plt.show()
```

```python
# Scatter plot - Deneyim vs Maaş
plt.figure(figsize=(10, 6))
plt.scatter(df['deneyim_yil'], df['maas'], alpha=0.5, c='green')
plt.title('Deneyim ve Maaş İlişkisi', fontsize=14)
plt.xlabel('Deneyim (Yıl)', fontsize=12)
plt.ylabel('Maaş (TL)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Subplot ile birden fazla grafik
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Grafik 1: Departman dağılımı
dept_counts = df['departman'].value_counts()
axes[0, 0].bar(dept_counts.index, dept_counts.values, color='skyblue')
axes[0, 0].set_title('Departman Dağılımı')
axes[0, 0].tick_params(axis='x', rotation=45)

# Grafik 2: Cinsiyet dağılımı
cinsiyet_counts = df['cinsiyet'].value_counts()
axes[0, 1].pie(cinsiyet_counts.values, labels=cinsiyet_counts.index,
               autopct='%1.1f%%', startangle=90)
axes[0, 1].set_title('Cinsiyet Dağılımı')

# Grafik 3: Maaş dağılımı
axes[1, 0].hist(df['maas'], bins=15, color='coral', edgecolor='black')
axes[1, 0].set_title('Maaş Dağılımı')
axes[1, 0].set_xlabel('Maaş')

# Grafik 4: Yaş vs Maaş
axes[1, 1].scatter(df['yas'], df['maas'], alpha=0.5)
axes[1, 1].set_title('Yaş ve Maaş İlişkisi')
axes[1, 1].set_xlabel('Yaş')
axes[1, 1].set_ylabel('Maaş')

plt.tight_layout()
plt.show()
```

```python
# Gelişmiş özelleştirme
plt.style.use('seaborn-v0_8-darkgrid')

fig, ax = plt.subplots(figsize=(12, 6))

# Departmanlara göre ortalama maaş
dept_maas = df.groupby('departman')['maas'].mean().sort_values(ascending=True)

bars = ax.barh(dept_maas.index, dept_maas.values, color='teal', alpha=0.8)

# Her çubuk için değer etiketi
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax.text(width, bar.get_y() + bar.get_height()/2,
            f'{width:,.0f} TL',
            ha='left', va='center', fontsize=10, fontweight='bold')

ax.set_xlabel('Ortalama Maaş (TL)', fontsize=12, fontweight='bold')
ax.set_title('Departmanlara Göre Ortalama Maaş', fontsize=14, fontweight='bold', pad=20)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Tema ayarlama
sns.set_theme(style="whitegrid")
sns.set_palette("husl")
```

```python
# Count plot - Kategorik değişken dağılımı
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='departman', order=df['departman'].value_counts().index)
plt.title('Departmanlara Göre Çalışan Sayısı', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Box plot - Maaş dağılımı
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='departman', y='maas')
plt.title('Departmanlara Göre Maaş Dağılımı', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='deneyim_yil', y='maas', scatter_kws={'alpha':0.5})
plt.title('Deneyim ve Maaş İlişkisi (Regresyon Çizgisi ile)', fontsize=14)
plt.xlabel('Deneyim (Yıl)')
plt.ylabel('Maaş (TL)')
plt.tight_layout()
plt.show()
```

```python
# Pair plot - Birden fazla değişken arası ilişki
numeric_cols = ['yas', 'deneyim_yil', 'maas', 'performans_notu']
sns.pairplot(df[numeric_cols + ['cinsiyet']], hue='cinsiyet',
             diag_kind='kde', palette='Set2')
plt.suptitle('Çoklu Değişken İlişkisi Analizi', y=1.02)
plt.tight_layout()
plt.show()
```

```python

# Correlation heatmap
plt.figure(figsize=(10, 8))
correlation = df[numeric_cols].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0,
            fmt='.2f', square=True, linewidths=1)
plt.title('Korelasyon Matrisi', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Violin plot - Dağılım yoğunluğu
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='departman', y='maas', hue='cinsiyet', split=True)
plt.title('Departman ve Cinsiyete Göre Maaş Dağılımı', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Swarm plot - Tüm veri noktaları
plt.figure(figsize=(12, 6))
sns.swarmplot(data=df[df['departman'].isin(['IT', 'HR', 'Finans'])],
              x='departman', y='maas', hue='cinsiyet', size=4)
plt.title('Seçili Departmanlarda Maaş Dağılımı (Detaylı)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python

# Bar plot with error bars
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='departman', y='maas', hue='egitim_seviyesi',
            ci=95, capsize=0.1)
plt.title('Departman ve Eğitim Seviyesine Göre Ortalama Maaş', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Eğitim Seviyesi', bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.dates as mdates

# İşe giriş tarihlerine göre analiz
df['ise_giris_tarihi'] = pd.to_datetime(df['ise_giris_tarihi']) # Ensure datetime type
df['ise_giris_yil'] = df['ise_giris_tarihi'].dt.year
df['ise_giris_ay'] = df['ise_giris_tarihi'].dt.to_period('M')

# Aylık işe alım trendi
aylik_ise_alim = df.groupby('ise_giris_ay').size().reset_index(name='sayi')
# 'ise_giris_ay' sütununu düzgün grafik çizimi ve biçimlendirme için datetime nesnelerine dönüştürme
aylik_ise_alim['ise_giris_ay'] = aylik_ise_alim['ise_giris_ay'].dt.to_timestamp()

plt.figure(figsize=(15, 6))
sns.lineplot(data=aylik_ise_alim, x='ise_giris_ay', y='sayi',
             marker='o', linewidth=2)
plt.title('Aylık İşe Alım Trendi', fontsize=14)
plt.xlabel('Tarih')
plt.ylabel('İşe Alım Sayısı')

# X ekseni için tarih biçimlendirmesini iyileştirmek üzere matplotlib.dates kullanma
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.YearLocator()) # Ana etiketleri yıllara ayarlama
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y')) # Yılları dört haneli tam sayı olarak biçimlendirme
ax.xaxis.set_minor_locator(mdates.MonthLocator(interval=3)) # İkincil etiketleri 3 aylık aralıklarla ayarlama
ax.xaxis.set_minor_formatter(mdates.DateFormatter('%Y-%m')) # Ayları YYYY-MM olarak biçimlendirme


plt.xticks(rotation=45) # Genel okunabilirlik için döndürmeyi koruma
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Facet grid - Çoklu kategoriye göre ayrı grafikler
g = sns.FacetGrid(df, col='departman', col_wrap=3, height=4, aspect=1.2)
g.map(sns.histplot, 'maas', kde=True, bins=15)
g.set_titles('{col_name}')
g.set_axis_labels('Maaş (TL)', 'Sıklık')
plt.tight_layout()
plt.show()
```
