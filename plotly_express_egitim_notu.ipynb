## 1. Plotly Express'e Giri≈ü ve Kurulum

### 1.1 Plotly Express Nedir?

**Plotly Express** (`plotly.express`), Plotly k√ºt√ºphanesinin y√ºksek seviyeli aray√ºz√ºd√ºr. Tek satƒ±r kodla profesyonel, interaktif grafikler olu≈üturmayƒ± saƒülar.

| √ñzellik | Matplotlib | Seaborn | Plotly Express |
|---------|-----------|---------|----------------|
| √áƒ±ktƒ± t√ºr√º | Statik g√∂rsel | Statik g√∂rsel | **ƒ∞nteraktif HTML** |
| Hover bilgisi | ‚ùå | ‚ùå | ‚úÖ Otomatik |
| Yakƒ±nla≈ütƒ±rma | ‚ùå | ‚ùå | ‚úÖ |
| Animasyon | ‚ùå | ‚ùå | ‚úÖ |
| Dashboard desteƒüi | Sƒ±nƒ±rlƒ± | Sƒ±nƒ±rlƒ± | ‚úÖ (Dash ile) |
| DataFrame entegrasyonu | Sƒ±nƒ±rlƒ± | ƒ∞yi | ‚úÖ M√ºkemmel |
| Dƒ±≈üa aktarma | PNG, PDF | PNG, PDF | **HTML, PNG, PDF, SVG** |

### 1.2 Temel Arg√ºmanlar

Plotly Express fonksiyonlarƒ±nda sƒ±klƒ±kla kullanƒ±lan ortak parametreler:

| Parametre | A√ßƒ±klama |
|-----------|----------|
| `data_frame` | Pandas DataFrame |
| `x`, `y` | Eksen deƒüi≈ükenleri |
| `color` | Renk bazƒ±nda gruplama (Seaborn'daki `hue`) |
| `size` | Nokta boyutunu belirleyen deƒüi≈üken |
| `hover_data` | Fareyle √ºzerine gelince g√∂sterilecek ek s√ºtunlar |
| `facet_col` / `facet_row` | √áoklu panel olu≈üturma |
| `animation_frame` | Animasyon olu≈üturma |
| `title` | Grafik ba≈ülƒ±ƒüƒ± |
| `labels` | Eksen ve legend etiketlerini √∂zelle≈ütirme |
| `color_discrete_sequence` | Kategorik renk paleti |
| `color_continuous_scale` | S√ºrekli renk skalasƒ± |
| `template` | Tema (`plotly`, `plotly_dark`, `seaborn`, `simple_white` vb.) |

```python
# Kurulum (gerekirse)
# !pip install plotly --quiet

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np

print('Plotly s√ºr√ºm:', px.__version__ if hasattr(px, '__version__') else 'Y√ºkl√º')
```

### 1.3 HR Veri Seti Olu≈üturma

```python
# ===== KAPSAMLI HR VERƒ∞ SETƒ∞ =====
np.random.seed(42)
n = 600

departmanlar = np.random.choice(
    ['M√ºhendislik', 'Pazarlama', 'ƒ∞K', 'Finans', 'Operasyon', 'Satƒ±≈ü'], n,
    p=[0.25, 0.18, 0.10, 0.15, 0.15, 0.17])

dept_maas = {'M√ºhendislik': 20000, 'Finans': 18000, 'Satƒ±≈ü': 15000,
             'Pazarlama': 16000, 'ƒ∞K': 14000, 'Operasyon': 14500}

hr = pd.DataFrame({
    'Calisan_ID': [f'EMP-{i:04d}' for i in range(1, n+1)],
    'Ad': [f'√áalƒ±≈üan_{i}' for i in range(1, n+1)],
    'Departman': departmanlar,
    'Pozisyon': np.random.choice(['Junior', 'Mid', 'Senior', 'Lead', 'M√ºd√ºr'], n,
                                  p=[0.25, 0.30, 0.25, 0.12, 0.08]),
    'Cinsiyet': np.random.choice(['Erkek', 'Kadƒ±n'], n, p=[0.54, 0.46]),
    'Yas': np.random.normal(34, 8, n).clip(22, 60).astype(int),
    'Kidem_Yil': np.random.exponential(4.5, n).clip(0.5, 22).round(1),
    'Egitim_Duzeyi': np.random.choice(['Lisans', 'Y√ºksek Lisans', 'Doktora'], n,
                                       p=[0.55, 0.35, 0.10]),
    'Calisma_Sekli': np.random.choice(['Ofis', 'Uzaktan', 'Hibrit'], n,
                                       p=[0.35, 0.30, 0.35]),
    'Performans': np.random.choice([1, 2, 3, 4, 5], n,
                                    p=[0.04, 0.12, 0.34, 0.32, 0.18]),
    'Memnuniyet': np.random.uniform(2, 10, n).round(1),
    'Egitim_Saati': np.random.normal(40, 15, n).clip(5, 100).astype(int),
    'Ayrilma': np.random.choice(['Aktif', 'Ayrƒ±ldƒ±'], n, p=[0.83, 0.17]),
})

# Maa≈üƒ± departman + kƒ±dem + pozisyon bazƒ±nda olu≈ütur
poz_carpan = {'Junior': 1.0, 'Mid': 1.25, 'Senior': 1.55, 'Lead': 1.85, 'M√ºd√ºr': 2.3}
hr['Maas'] = hr.apply(
    lambda r: int(dept_maas[r['Departman']] * poz_carpan[r['Pozisyon']]
                  + r['Kidem_Yil'] * 400
                  + np.random.normal(0, 1500)), axis=1)
hr['Maas'] = hr['Maas'].clip(9000, 65000)

# Aylƒ±k trend verisi
aylar_list = pd.date_range('2023-01', periods=24, freq='M')
trend = pd.DataFrame({
    'Tarih': np.tile(aylar_list, 3),
    'Departman': np.repeat(['M√ºhendislik', 'Pazarlama', 'Satƒ±≈ü'], 24),
    'Ise_Alim': np.concatenate([
        np.random.poisson(10, 24),
        np.random.poisson(6, 24),
        np.random.poisson(8, 24)
    ]),
    'Ayrilma_Sayisi': np.concatenate([
        np.random.poisson(3, 24),
        np.random.poisson(2, 24),
        np.random.poisson(4, 24)
    ])
})

# Yƒ±l bazƒ±nda √∂zet
yil_ozet = hr.groupby(['Departman', 'Pozisyon']).agg(
    Ortalama_Maas=('Maas', 'mean'),
    Calisan_Sayisi=('Calisan_ID', 'count'),
    Ort_Memnuniyet=('Memnuniyet', 'mean'),
    Ort_Performans=('Performans', 'mean')
).reset_index()

print(f'HR veri seti: {hr.shape[0]} √ßalƒ±≈üan, {hr.shape[1]} s√ºtun')
print(f'Trend verisi: {trend.shape[0]} satƒ±r')
hr.head(8)
```

---
## 2. √áizgi Grafiƒüi (`px.line`)

Zaman serisi verileri ve trendleri g√∂rselle≈ütirmek i√ßin kullanƒ±lƒ±r. `color` parametresi ile birden fazla seri aynƒ± grafikte g√∂sterilebilir.

```python
# HR √ñrnek: Aylƒ±k i≈üe alƒ±m trendi (departman bazƒ±nda)
fig = px.line(trend, x='Tarih', y='Ise_Alim',
              color='Departman',
              markers=True,
              title='Aylƒ±k ƒ∞≈üe Alƒ±m Trendi (2023-2024)',
              labels={'Ise_Alim': 'ƒ∞≈üe Alƒ±nan Ki≈üi',
                      'Tarih': 'Ay'},
              template='plotly_white',
              color_discrete_sequence=px.colors.qualitative.Set2)

fig.update_layout(hovermode='x unified',
                  legend_title='Departman',
                  font=dict(size=12))
fig.show()
```

```python
# √áoklu eksen: ƒ∞≈üe alƒ±m vs Ayrƒ±lma (M√ºhendislik departmanƒ±)
muh_trend = trend[trend['Departman'] == 'M√ºhendislik'].copy()

fig = go.Figure()
fig.add_trace(go.Scatter(x=muh_trend['Tarih'], y=muh_trend['Ise_Alim'],
                         name='ƒ∞≈üe Alƒ±m', mode='lines+markers',
                         line=dict(color='#2ECC71', width=2)))
fig.add_trace(go.Scatter(x=muh_trend['Tarih'], y=muh_trend['Ayrilma_Sayisi'],
                         name='Ayrƒ±lma', mode='lines+markers',
                         line=dict(color='#E74C3C', width=2, dash='dash')))

fig.update_layout(title='M√ºhendislik: ƒ∞≈üe Alƒ±m vs Ayrƒ±lma Trendi',
                  xaxis_title='Tarih', yaxis_title='Ki≈üi Sayƒ±sƒ±',
                  template='plotly_white', hovermode='x unified')
fig.show()
```

---
## 3. √áubuk Grafiƒüi (`px.bar`)

Kategorik kar≈üƒ±la≈ütƒ±rmalar i√ßin kullanƒ±lƒ±r. `barmode` parametresi ile `'group'` (yan yana) veya `'stack'` (yƒ±ƒüƒ±lmƒ±≈ü) d√ºzenleri se√ßilebilir.

```python
# HR √ñrnek: Departman bazƒ±nda √ßalƒ±≈üan sayƒ±sƒ± (cinsiyete g√∂re)
dept_cinsiyet = hr.groupby(['Departman', 'Cinsiyet']).size().reset_index(name='Sayi')

fig = px.bar(dept_cinsiyet, x='Departman', y='Sayi',
             color='Cinsiyet', barmode='group',
             title='Departman ve Cinsiyete G√∂re √áalƒ±≈üan Sayƒ±sƒ±',
             labels={'Sayi': '√áalƒ±≈üan Sayƒ±sƒ±'},
             color_discrete_map={'Erkek': '#3498DB', 'Kadƒ±n': '#E74C3C'},
             template='plotly_white',
             text_auto=True)

fig.update_traces(textposition='outside')
fig.update_layout(xaxis_categoryorder='total descending')
fig.show()
```

```python
# Yƒ±ƒüƒ±lmƒ±≈ü √ßubuk grafik: Pozisyon daƒüƒ±lƒ±mƒ±
poz_dept = hr.groupby(['Departman', 'Pozisyon']).size().reset_index(name='Sayi')

fig = px.bar(poz_dept, x='Departman', y='Sayi',
             color='Pozisyon', barmode='stack',
             title='Departman Bazƒ±nda Pozisyon Daƒüƒ±lƒ±mƒ± (Yƒ±ƒüƒ±lmƒ±≈ü)',
             labels={'Sayi': '√áalƒ±≈üan Sayƒ±sƒ±'},
             color_discrete_sequence=px.colors.qualitative.Pastel,
             template='plotly_white')

fig.update_layout(xaxis_categoryorder='total descending')
fig.show()
```

---
## 4. Daƒüƒ±lƒ±m Grafiƒüi (`px.scatter`)

ƒ∞ki s√ºrekli deƒüi≈üken arasƒ±ndaki ili≈ükiyi g√∂rselle≈ütirir. Plotly Express'te `size`, `color`, `symbol`, `hover_data` gibi parametrelerle √ßok boyutlu veri tek grafikte g√∂sterilebilir.

```python
# HR √ñrnek: Kƒ±dem vs Maa≈ü (4 boyutlu: renk=departman, boyut=performans)
fig = px.scatter(hr, x='Kidem_Yil', y='Maas',
                 color='Departman',
                 size='Performans',
                 hover_data=['Ad', 'Pozisyon', 'Yas', 'Memnuniyet'],
                 title='Kƒ±dem - Maa≈ü ƒ∞li≈ükisi (Departman ve Performans)',
                 labels={'Kidem_Yil': 'Kƒ±dem (Yƒ±l)',
                         'Maas': 'Aylƒ±k Maa≈ü (TL)',
                         'Performans': 'Performans Puanƒ±'},
                 opacity=0.6,
                 template='plotly_white',
                 color_discrete_sequence=px.colors.qualitative.Set1)

fig.update_layout(legend_title='Departman')
fig.show()
```

```python
# Trendline (regresyon √ßizgisi) ekleme
fig = px.scatter(hr, x='Kidem_Yil', y='Maas',
                 color='Cinsiyet',
                 trendline='ols',
                 title='Kƒ±dem - Maa≈ü Regresyon Analizi (Cinsiyete G√∂re)',
                 labels={'Kidem_Yil': 'Kƒ±dem (Yƒ±l)', 'Maas': 'Maa≈ü (TL)'},
                 color_discrete_map={'Erkek': '#2E86C1', 'Kadƒ±n': '#C0392B'},
                 opacity=0.4,
                 template='plotly_white')
fig.show()
```

---
## 5. Pasta Grafiƒüi (`px.pie`)

Oransal daƒüƒ±lƒ±mlarƒ± g√∂sterir. `hole` parametresi ile halka grafiƒüe (donut) d√∂n√º≈üt√ºr√ºlebilir.

```python
# HR √ñrnek: Departman √ßalƒ±≈üan oranƒ±
dept_sayi = hr['Departman'].value_counts().reset_index()
dept_sayi.columns = ['Departman', 'Sayi']

fig = px.pie(dept_sayi, values='Sayi', names='Departman',
             title='Departman Bazƒ±nda √áalƒ±≈üan Oranƒ±',
             color_discrete_sequence=px.colors.qualitative.Set2,
             hole=0,  # 0 = pasta, 0.4 = donut
             template='plotly_white')

fig.update_traces(textposition='inside', textinfo='percent+label',
                  hoverinfo='label+value+percent')
fig.show()
```

---
## 6. Halka Grafiƒüi (Donut Chart)

`px.pie` fonksiyonuna `hole` parametresi eklenerek olu≈üturulur. Ortadaki bo≈üluƒüa √∂zet bilgi eklenebilir.

```python
# HR √ñrnek: Ayrƒ±lma durumu halka grafiƒüi
ayrilma_sayi = hr['Ayrilma'].value_counts().reset_index()
ayrilma_sayi.columns = ['Durum', 'Sayi']

fig = px.pie(ayrilma_sayi, values='Sayi', names='Durum',
             title='√áalƒ±≈üan Durumu Daƒüƒ±lƒ±mƒ±',
             hole=0.45,
             color_discrete_map={'Aktif': '#2ECC71', 'Ayrƒ±ldƒ±': '#E74C3C'})

fig.update_traces(textinfo='percent+value',
                  marker=dict(line=dict(color='white', width=2)))
fig.add_annotation(text=f'{len(hr)}<br>Toplam',
                   showarrow=False, font_size=18,
                   font_color='#2C3E50')
fig.show()
```

---
## 7. Histogram (`px.histogram`)

S√ºrekli deƒüi≈ükenlerin frekans daƒüƒ±lƒ±mƒ±nƒ± g√∂sterir. `marginal` parametresi ile kenarlara ek grafikler eklenebilir: `'rug'`, `'box'`, `'violin'`, `'histogram'`.

```python
# HR √ñrnek: Maa≈ü daƒüƒ±lƒ±mƒ± (departman bazƒ±nda)
fig = px.histogram(hr, x='Maas',
                   color='Departman',
                   nbins=30,
                   marginal='box',
                   title='Departman Bazƒ±nda Maa≈ü Daƒüƒ±lƒ±mƒ±',
                   labels={'Maas': 'Aylƒ±k Maa≈ü (TL)', 'count': 'Frekans'},
                   opacity=0.7,
                   barmode='overlay',
                   template='plotly_white',
                   color_discrete_sequence=px.colors.qualitative.Set2)

fig.show()
```

```python
# Ya≈ü daƒüƒ±lƒ±mƒ± (marginal='violin')
fig = px.histogram(hr, x='Yas',
                   color='Cinsiyet',
                   nbins=25,
                   marginal='violin',
                   title='Cinsiyete G√∂re Ya≈ü Daƒüƒ±lƒ±mƒ±',
                   labels={'Yas': 'Ya≈ü'},
                   color_discrete_map={'Erkek': '#3498DB', 'Kadƒ±n': '#E74C3C'},
                   opacity=0.6, barmode='overlay',
                   template='plotly_white')
fig.show()
```

---
## 8. Kutu Grafiƒüi (`px.box`)

Be≈ü istatistiksel √∂zetle (min, Q1, medyan, Q3, max) daƒüƒ±lƒ±mƒ± g√∂sterir. `points` parametresi ile bireysel noktalar da eklenebilir: `'all'`, `'outliers'`, `'suspectedoutliers'`, `False`.

```python
# HR √ñrnek: Pozisyon bazƒ±nda maa≈ü daƒüƒ±lƒ±mƒ±
fig = px.box(hr, x='Pozisyon', y='Maas',
             color='Cinsiyet',
             points='outliers',
             title='Pozisyon ve Cinsiyete G√∂re Maa≈ü Daƒüƒ±lƒ±mƒ±',
             labels={'Maas': 'Aylƒ±k Maa≈ü (TL)'},
             color_discrete_map={'Erkek': '#2E86C1', 'Kadƒ±n': '#C0392B'},
             category_orders={'Pozisyon': ['Junior', 'Mid', 'Senior', 'Lead', 'M√ºd√ºr']},
             template='plotly_white')

fig.update_layout(boxmode='group')
fig.show()
```

---
## 9. Keman Grafiƒüi (`px.violin`)

Kutu grafiƒüi + √ßekirdek yoƒüunluk tahmini (KDE) birle≈üimi. `box=True` ile i√ß kutu grafiƒüi, `points='all'` ile t√ºm noktalar g√∂sterilebilir.

```python
# HR √ñrnek: √áalƒ±≈üma ≈üekline g√∂re memnuniyet daƒüƒ±lƒ±mƒ±
fig = px.violin(hr, x='Calisma_Sekli', y='Memnuniyet',
                color='Calisma_Sekli',
                box=True,
                points='all',
                title='√áalƒ±≈üma ≈ûekline G√∂re Memnuniyet Daƒüƒ±lƒ±mƒ±',
                labels={'Memnuniyet': 'Memnuniyet Puanƒ± (1-10)',
                        'Calisma_Sekli': '√áalƒ±≈üma ≈ûekli'},
                color_discrete_sequence=['#3498DB', '#2ECC71', '#F39C12'],
                template='plotly_white')

fig.update_traces(meanline_visible=True)
fig.show()
```

---
## 10. ≈ûerit Grafiƒüi (`px.strip`)

Her g√∂zlemi bireysel nokta olarak g√∂sterir. Seaborn'daki `stripplot()` fonksiyonunun Plotly kar≈üƒ±lƒ±ƒüƒ±dƒ±r.

```python
# HR √ñrnek: Departman bazƒ±nda eƒüitim saati
fig = px.strip(hr, x='Departman', y='Egitim_Saati',
               color='Performans',
               hover_data=['Ad', 'Pozisyon'],
               title='Departman Bazƒ±nda Eƒüitim Saatleri (Her √áalƒ±≈üan)',
               labels={'Egitim_Saati': 'Eƒüitim Saati (Yƒ±llƒ±k)'},
               template='plotly_white')
fig.show()
```

## 11. Isƒ± Haritasƒ± (`px.imshow` / `px.density_heatmap`)

- `px.imshow`: Korelasyon matrisi gibi hazƒ±r 2D dizileri g√∂stermek i√ßin
- `px.density_heatmap`: ƒ∞ki deƒüi≈ükenin yoƒüunluk haritasƒ± i√ßin

```python
# HR √ñrnek: Korelasyon ƒ±sƒ± haritasƒ±
sayisal = ['Yas', 'Kidem_Yil', 'Maas', 'Performans', 'Egitim_Saati', 'Memnuniyet']
kor = hr[sayisal].corr().round(2)

fig = px.imshow(kor, text_auto=True,
                color_continuous_scale='RdYlBu_r',
                zmin=-1, zmax=1,
                title='HR Deƒüi≈ükenleri Korelasyon Matrisi',
                aspect='auto')

fig.update_layout(width=650, height=550)
fig.show()
```

```python
# Yoƒüunluk ƒ±sƒ± haritasƒ±: Ya≈ü vs Kƒ±dem
fig = px.density_heatmap(hr, x='Yas', y='Kidem_Yil',
                         marginal_x='histogram',
                         marginal_y='histogram',
                         title='Ya≈ü ve Kƒ±dem Yoƒüunluk Haritasƒ±',
                         labels={'Yas': 'Ya≈ü', 'Kidem_Yil': 'Kƒ±dem (Yƒ±l)'},
                         template='plotly_white')
# 'Blues' renk skalasƒ±nƒ± ana heatmap izine ayrƒ± olarak uygula
# px.density_heatmap'in ilk izi genellikle ana heatmap'tir.
if len(fig.data) > 0 and hasattr(fig.data[0], 'colorscale'):
    fig.data[0].colorscale = 'Blues'
fig.show()
```

---
## 12. Aƒüa√ß Haritasƒ± (`px.treemap`)

Hiyerar≈üik verileri i√ß i√ße ge√ßmi≈ü dikd√∂rtgenler olarak g√∂sterir. `path` parametresi ile hiyerar≈üi seviyesi belirlenir. HR'da organizasyon yapƒ±sƒ±nƒ± g√∂rselle≈ütirmek i√ßin idealdir.

## 15. Animasyonlu Grafikler (`animation_frame`)

Plotly Express'in en g√º√ßl√º √∂zelliklerinden biri **animasyon** desteƒüidir. `animation_frame` parametresi ile zaman bazlƒ± animasyonlar olu≈üturulabilir. Bu √∂zellik Matplotlib ve Seaborn'da doƒürudan mevcut deƒüildir.

```python
# HR √ñrnek: Yƒ±l bazƒ±nda departman maa≈ü daƒüƒ±lƒ±mƒ± animasyonu
# Sim√ºlasyon verisi olu≈ütur (3 yƒ±l)
np.random.seed(42)
anim_frames = []
for yil in [2023, 2024, 2025]:
    for dept in hr['Departman'].unique():
        mask = hr['Departman'] == dept
        n_dept = mask.sum()
        artis = (yil - 2023) * 1200
        anim_frames.append(pd.DataFrame({
            'Departman': [dept] * n_dept,
            'Maas': hr.loc[mask, 'Maas'] + artis + np.random.normal(0, 500, n_dept),
            'Kidem_Yil': hr.loc[mask, 'Kidem_Yil'],
            'Performans': hr.loc[mask, 'Performans'],
            'Yil': [str(yil)] * n_dept
        }))

anim_data = pd.concat(anim_frames, ignore_index=True)

fig = px.scatter(anim_data, x='Kidem_Yil', y='Maas',
                 color='Departman', size='Performans',
                 animation_frame='Yil',
                 title='Yƒ±l Bazƒ±nda Kƒ±dem-Maa≈ü Deƒüi≈üimi (Animasyonlu)',
                 labels={'Kidem_Yil': 'Kƒ±dem (Yƒ±l)', 'Maas': 'Maa≈ü (TL)'},
                 range_y=[5000, 70000],
                 range_x=[0, 25],
                 opacity=0.5,
                 template='plotly_white')

fig.update_layout(width=800, height=500)
fig.show()
```

## 17. Radar / Kutup Grafiƒüi (`px.line_polar`)

Birden fazla metriƒüi tek bir dairesel grafikte kar≈üƒ±la≈ütƒ±rmak i√ßin kullanƒ±lƒ±r. Departman performans kar≈üƒ±la≈ütƒ±rmasƒ± gibi senaryolarda √ßok etkilidir.

```python
# HR √ñrnek: Departman bazƒ±nda √ßoklu metrik radar grafiƒüi
from sklearn.preprocessing import MinMaxScaler

radar_df = hr.groupby('Departman').agg(
    Ort_Maas=('Maas', 'mean'),
    Ort_Performans=('Performans', 'mean'),
    Ort_Memnuniyet=('Memnuniyet', 'mean'),
    Ort_Egitim=('Egitim_Saati', 'mean'),
    Ort_Kidem=('Kidem_Yil', 'mean')
).reset_index()

# 0-10 arasƒ± normalize et
metrikler = ['Ort_Maas', 'Ort_Performans', 'Ort_Memnuniyet', 'Ort_Egitim', 'Ort_Kidem']
for col in metrikler:
    radar_df[col] = ((radar_df[col] - radar_df[col].min()) /
                     (radar_df[col].max() - radar_df[col].min()) * 10).round(1)

# Long format'a √ßevir
radar_long = radar_df.melt(id_vars='Departman', value_vars=metrikler,
                            var_name='Metrik', value_name='Skor')
radar_long['Metrik'] = radar_long['Metrik'].map({
    'Ort_Maas': 'Maa≈ü', 'Ort_Performans': 'Performans',
    'Ort_Memnuniyet': 'Memnuniyet', 'Ort_Egitim': 'Eƒüitim',
    'Ort_Kidem': 'Kƒ±dem'})

fig = px.line_polar(radar_long, r='Skor', theta='Metrik',
                    color='Departman', line_close=True,
                    title='Departman Performans Radar Grafiƒüi',
                    template='plotly_white',
                    color_discrete_sequence=px.colors.qualitative.Set2)

fig.update_traces(fill='toself', opacity=0.5)
fig.update_layout(polar=dict(radialaxis=dict(range=[0, 10])))
fig.show()
```

## 20. Alt Grafikler (Subplots) ‚Äì `make_subplots`

Plotly'de `make_subplots` fonksiyonu ile birden fazla grafik tek bir fig√ºrde birle≈ütirilebilir. `specs` parametresi ile grafik t√ºrleri (scatter, pie vb.) belirlenebilir.

```python
# HR √ñrnek: 2x2 subplot dashboard
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('Maa≈ü Daƒüƒ±lƒ±mƒ±', 'Departman √áalƒ±≈üan Sayƒ±sƒ±',
                    'Kƒ±dem vs Maa≈ü', 'Performans Daƒüƒ±lƒ±mƒ±'),
    specs=[[{'type': 'histogram'}, {'type': 'bar'}],
           [{'type': 'scatter'}, {'type': 'pie'}]]
)

# 1) Histogram - Maa≈ü
fig.add_trace(
    go.Histogram(x=hr['Maas'], nbinsx=25, name='Maa≈ü',
                 marker_color='#3498DB', opacity=0.8),
    row=1, col=1)

# 2) Bar - Departman sayƒ±larƒ±
dept_counts = hr['Departman'].value_counts()
fig.add_trace(
    go.Bar(x=dept_counts.index, y=dept_counts.values, name='√áalƒ±≈üan',
           marker_color=px.colors.qualitative.Set2[:len(dept_counts)]),
    row=1, col=2)

# 3) Scatter - Kƒ±dem vs Maa≈ü
fig.add_trace(
    go.Scatter(x=hr['Kidem_Yil'], y=hr['Maas'], mode='markers',
               name='Kƒ±dem-Maa≈ü',
               marker=dict(color='#9B59B6', size=4, opacity=0.4)),
    row=2, col=1)

# 4) Pie - Performans daƒüƒ±lƒ±mƒ±
perf_counts = hr['Performans'].value_counts().sort_index()
fig.add_trace(
    go.Pie(labels=[f'Puan {p}' for p in perf_counts.index],
           values=perf_counts.values, name='Performans'),
    row=2, col=2)

fig.update_layout(
    height=700, width=900,
    title_text='HR √ñzet Dashboard (make_subplots)',
    title_font_size=18,
    showlegend=False,
    template='plotly_white')
fig.show()
```
