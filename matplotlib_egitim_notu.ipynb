## 2. `plot()` Fonksiyonu ile Ã‡izgi GrafiÄŸi

`plot()` fonksiyonu, Matplotlib'in en temel ve Ã§ok yÃ¶nlÃ¼ komutudur. Tek eksen, iki eksen, tekli ve Ã§oklu veri setleri iÃ§in grafikler Ã§izebilir. Her x, y Ã§ifti iÃ§in isteÄŸe baÄŸlÄ± bir biÃ§im dizesi (format string) ile renk ve Ã§izgi tÃ¼rÃ¼ belirlenebilir.

### 2.1 Temel KullanÄ±m

```python
# HR Ã–rnek: AylÄ±k iÅŸe alÄ±m sayÄ±larÄ±
aylar = ["Oca", "Åžub", "Mar", "Nis", "May", "Haz",
         "Tem", "AÄŸu", "Eyl", "Eki", "Kas", "Ara"]
ise_alim = [12, 8, 15, 10, 18, 14, 20, 16, 12, 9, 11, 22]

plt.figure(figsize=(10, 5))
plt.plot(aylar, ise_alim, 'b-o', linewidth=2, markersize=6)
plt.title("AylÄ±k Ä°ÅŸe AlÄ±m Trendi", fontsize=14, fontweight="bold")
plt.xlabel("Aylar")
plt.ylabel("Ä°ÅŸe AlÄ±nan KiÅŸi SayÄ±sÄ±")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 2.2 BiÃ§im Dizesi (Format String)

Her x ve y Ã§ifti iÃ§in renk ve ÅŸekil belirten Ã¼Ã§Ã¼ncÃ¼ bir parametre eklenebilir. VarsayÄ±lan biÃ§im `"b-"` yani mavi dÃ¼z Ã§izgidir.

| Sembol | AÃ§Ä±klama | Sembol | AÃ§Ä±klama |
|--------|----------|--------|----------|
| `b--` | Mavi kesikli Ã§izgi | `rs` | KÄ±rmÄ±zÄ± kare |
| `go` | YeÅŸil daire | `m^` | Magenta Ã¼Ã§gen |
| `ro` | KÄ±rmÄ±zÄ± nokta | `k-` | Siyah dÃ¼z Ã§izgi |

**Eksen SÄ±nÄ±rlarÄ± ve Ã–zelleÅŸtirme FonksiyonlarÄ±:**
- `axis([xmin, xmax, ymin, ymax])`: Her iki eksenin sÄ±nÄ±rÄ±nÄ± tek komutla belirler
- `xlim(min, max)` / `ylim(min, max)`: Tek eksen sÄ±nÄ±rÄ± ayarlar
- `title()`: GrafiÄŸe baÅŸlÄ±k ekler
- `xlabel()` / `ylabel()`: Eksenlere etiket ekler
- `grid(True)`: GrafiÄŸe Ä±zgara ekler
- `text(x, y, metin)`: Belirtilen konuma yazÄ± ekler

### 2.3 Ã‡oklu Ã‡izgi GrafiÄŸi

```python
# HR Ã–rnek: Departman bazÄ±nda Ã§alÄ±ÅŸan sayÄ±sÄ± trendi
yillar = [2020, 2021, 2022, 2023, 2024, 2025]
muhendislik = [45, 52, 58, 65, 72, 80]
pazarlama = [30, 35, 38, 42, 45, 50]
ik_departmani = [15, 18, 20, 22, 25, 28]
finans = [20, 22, 24, 26, 28, 30]

plt.figure(figsize=(10, 6))
plt.plot(yillar, muhendislik, "b--o", label="MÃ¼hendislik", linewidth=2)
plt.plot(yillar, pazarlama, "rs-", label="Pazarlama", linewidth=2)
plt.plot(yillar, ik_departmani, "g^-", label="Ä°K", linewidth=2)
plt.plot(yillar, finans, "m*-", label="Finans", linewidth=2)

plt.title("Departman BazÄ±nda Ã‡alÄ±ÅŸan SayÄ±sÄ± Trendi", fontsize=14, fontweight="bold")
plt.xlabel("YÄ±l")
plt.ylabel("Ã‡alÄ±ÅŸan SayÄ±sÄ±")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### 2.4 Stil KullanÄ±mÄ± (ggplot)

`matplotlib.style` modÃ¼lÃ¼ ile grafiklere Ã¶zel gÃ¶rsel efektler eklenebilir. Ã–rneÄŸin `"ggplot"` stili, gri arka plan ve eksen Ã§izgileriyle profesyonel bir gÃ¶rÃ¼nÃ¼m sunar.

```python
# ggplot stili ile aynÄ± grafik
from matplotlib import style
style.use("ggplot")

plt.figure(figsize=(10, 5))
plt.plot(yillar, muhendislik, 'o-', label='MÃ¼hendislik', linewidth=2)
plt.plot(yillar, pazarlama, 's-', label='Pazarlama', linewidth=2)
plt.title("ggplot Stili ile Ã‡alÄ±ÅŸan Trendi", fontsize=14)
plt.xlabel("YÄ±l")
plt.ylabel("Ã‡alÄ±ÅŸan SayÄ±sÄ±")
plt.legend()
plt.tight_layout()
plt.show()

# Stili varsayÄ±lana geri dÃ¶ndÃ¼r
style.use('default')
```

## 3. Pasta GrafiÄŸi (Pie Chart)

Pasta grafiÄŸi, mutlak ve gÃ¶reli frekanslarÄ± gÃ¶rselleÅŸtirir. Daire dilimlerinden oluÅŸur ve her dilim bir kategoriyi temsil eder. `pie()` fonksiyonu ile Ã§izilir. Dilim boyutu, gÃ¶reli frekansa baÄŸlÄ±dÄ±r ve aÃ§Ä± ile belirlenir.

```python
# HR Ã–rnek: Departman daÄŸÄ±lÄ±mÄ±
departmanlar = ['MÃ¼hendislik', 'Pazarlama', 'Ä°K', 'Finans', 'Operasyon']
calisan_sayisi = [120, 80, 35, 45, 60]
renkler = ['#3498DB', '#E74C3C', '#2ECC71', '#F39C12', '#9B59B6']
patlama = (0.05, 0, 0, 0, 0)  # MÃ¼hendisliÄŸi vurgula

plt.figure(figsize=(8, 8))
plt.pie(calisan_sayisi, labels=departmanlar,
        colors=renkler, explode=patlama,
        autopct='%1.1f%%', shadow=True,
        textprops={'fontsize': 12})
plt.title("Departman BazÄ±nda Ã‡alÄ±ÅŸan DaÄŸÄ±lÄ±mÄ±",
          fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

## 4. Keman GrafiÄŸi (Violin Plot)

Keman grafiÄŸi, kutu grafiÄŸi (box plot) ve Ã§ekirdek yoÄŸunluk grafiÄŸinin (kernel density plot) birleÅŸimidir. `violinplot()` fonksiyonu ile Ã§izilir. GrafiÄŸin geniÅŸliÄŸinin fazla olduÄŸu noktalarda daha fazla gÃ¶zlem bulunur.

- `showmeans=True/False`: OrtalamayÄ± gÃ¶sterir/gizler
- `showextrema=True/False`: UÃ§ deÄŸerleri gÃ¶sterir/gizler

```python
# HR Ã–rnek: Departman bazÄ±nda maaÅŸ daÄŸÄ±lÄ±mÄ±
np.random.seed(42)
muhendis_maas = np.random.normal(18000, 3000, 100)
pazarlama_maas = np.random.normal(14000, 2500, 80)
ik_maas = np.random.normal(13000, 2000, 50)
finans_maas = np.random.normal(16000, 2800, 60)

plt.figure(figsize=(10, 6))
vp = plt.violinplot([muhendis_maas, pazarlama_maas, ik_maas, finans_maas],
                     showmeans=True, showextrema=True)

# Renklendirme
colors = ['#3498DB', '#E74C3C', '#2ECC71', '#F39C12']
for i, body in enumerate(vp['bodies']):
    body.set_facecolor(colors[i])
    body.set_alpha(0.7)

plt.xticks([1, 2, 3, 4], ["MÃ¼hendislik", "Pazarlama", "Ä°K", "Finans"])
plt.ylabel("AylÄ±k MaaÅŸ (TL)", fontsize=12)
plt.title("Departman BazÄ±nda MaaÅŸ DaÄŸÄ±lÄ±mÄ±",
          fontsize=14, fontweight="bold")
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

## 5. DaÄŸÄ±lÄ±m GrafiÄŸi (Scatter Plot)

DaÄŸÄ±lÄ±m grafiÄŸi, `scatter()` fonksiyonu ile Ã§izilir ve iki deÄŸiÅŸkenli (bivariate) verileri gÃ¶rÃ¼ntÃ¼lemek iÃ§in kullanÄ±lÄ±r. Her nokta, iki deÄŸiÅŸkenin deÄŸerlerini temsil eder. Bir deÄŸiÅŸken yatay, diÄŸeri dikey eksende gÃ¶sterilir.

```python
# HR Ã–rnek: Deneyim vs MaaÅŸ iliÅŸkisi
np.random.seed(42)
deneyim = np.random.uniform(0.5, 20, 150)
maas = deneyim * 1500 + np.random.normal(5000, 2000, 150)

# Departmanlara gÃ¶re renklendirme
departman = np.random.choice(['MÃ¼hendislik', 'Pazarlama', 'Ä°K', 'Finans'], 150)
renk_map = {'MÃ¼hendislik': '#3498DB', 'Pazarlama': '#E74C3C',
            'Ä°K': '#2ECC71', 'Finans': '#F39C12'}
renkler = [renk_map[d] for d in departman]

plt.figure(figsize=(10, 6))
for dept_adi, renk in renk_map.items():
    mask = departman == dept_adi
    plt.scatter(deneyim[mask], maas[mask], color=renk,
               label=dept_adi, alpha=0.6, s=60)

plt.title("Deneyim - MaaÅŸ Ä°liÅŸkisi (Departman BazÄ±nda)",
          fontsize=14, fontweight="bold")
plt.xlabel("Deneyim (YÄ±l)", fontsize=12)
plt.ylabel("AylÄ±k MaaÅŸ (TL)", fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 6. Histogram

Histogram, verileri gruplara ayÄ±rarak her grubun frekansÄ±nÄ± Ã§ubuk yÃ¼ksekliÄŸi ile gÃ¶sterir. `hist()` fonksiyonu ile Ã§izilir ve sÃ¼rekli deÄŸiÅŸkenlerin daÄŸÄ±lÄ±mÄ±nÄ± analiz etmek iÃ§in idealdir. Her Ã§ubuk, o aralÄ±ktaki deÄŸerlerin sayÄ±sÄ±nÄ± temsil eder.

```python
# HR Ã–rnek: Ã‡alÄ±ÅŸan yaÅŸ daÄŸÄ±lÄ±mÄ±
np.random.seed(42)
yaslar = np.random.normal(35, 8, 500)
yaslar = yaslar[(yaslar >= 20) & (yaslar <= 65)]  # GerÃ§ekÃ§i aralÄ±k

plt.figure(figsize=(10, 6))
plt.hist(yaslar, bins=20, color='#3498DB',
         edgecolor='white', alpha=0.8)

# Ortalama Ã§izgisi
plt.axvline(np.mean(yaslar), color='#E74C3C', linestyle='--',
            linewidth=2, label=f'Ortalama: {np.mean(yaslar):.1f}')

plt.title("Ã‡alÄ±ÅŸan YaÅŸ DaÄŸÄ±lÄ±mÄ±", fontsize=14, fontweight="bold")
plt.xlabel("YaÅŸ", fontsize=12)
plt.ylabel("Frekans", fontsize=12)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

### 6.1 Tek GÃ¶rselde Ã‡oklu Grafik (subplot)

`subplot()` fonksiyonu ile tek bir gÃ¶rsel Ã¼zerinde m Ã— n dÃ¼zeninde birden fazla grafik oluÅŸturulabilir. ÃœÃ§ parametre alÄ±r: satÄ±r sayÄ±sÄ±, sÃ¼tun sayÄ±sÄ± ve hÃ¼cre numarasÄ±. HÃ¼cre numaralarÄ± satÄ±r bazÄ±nda belirlenir.

`figure()` fonksiyonunun `figsize` parametresi gÃ¶rselin boyutlarÄ±nÄ±, `suptitle()` ise tÃ¼m alt grafiklere ortak Ã¼st baÅŸlÄ±k ekler.

```python
# HR Ã–rnek: FarklÄ± departmanlar iÃ§in yaÅŸ daÄŸÄ±lÄ±mÄ±
np.random.seed(42)
muhendis_yas = np.random.normal(32, 6, 120)
pazarlama_yas = np.random.normal(30, 5, 80)
ik_yas = np.random.normal(35, 7, 50)
finans_yas = np.random.normal(38, 6, 45)

fig = plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(muhendis_yas, bins=12, color='#3498DB', edgecolor='white')
plt.title('MÃ¼hendislik')
plt.xlabel('YaÅŸ')

plt.subplot(2, 2, 2)
plt.hist(pazarlama_yas, bins=12, color='#E74C3C', edgecolor='white')
plt.title('Pazarlama')
plt.xlabel('YaÅŸ')

plt.subplot(2, 2, 3)
plt.hist(ik_yas, bins=12, color='#2ECC71', edgecolor='white')
plt.title('Ä°K')
plt.xlabel('YaÅŸ')

plt.subplot(2, 2, 4)
plt.hist(finans_yas, bins=12, color='#F39C12', edgecolor='white')
plt.title('Finans')
plt.xlabel('YaÅŸ')

plt.suptitle('Departman BazÄ±nda YaÅŸ DaÄŸÄ±lÄ±mÄ±',
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 7. Ã‡ubuk Grafik (Bar Chart)

Ã‡ubuk grafik, kategorik deÄŸiÅŸkenlerin frekanslarÄ±nÄ± dikdÃ¶rtgen Ã§ubuklar ile gÃ¶sterir. `bar()` fonksiyonu dikey, `barh()` fonksiyonu ise yatay Ã§ubuk grafik oluÅŸturur. YÄ±ÄŸÄ±lmÄ±ÅŸ (stacked) Ã§ubuk grafikler de Ã§izilebilir.

```python
# HR Ã–rnek: Departman bazÄ±nda performans puanlarÄ±
dept = ['MÃ¼hendislik', 'Pazarlama', 'Ä°K', 'Finans', 'Operasyon']
ortalama_puan = [4.2, 3.8, 4.0, 3.5, 3.9]
renkler = ['#2ECC71', '#3498DB', '#E74C3C', '#F39C12', '#9B59B6']

plt.figure(figsize=(10, 6))
bars = plt.bar(dept, ortalama_puan, color=renkler, edgecolor='white',
               linewidth=1.5)

# Her Ã§ubuÄŸun Ã¼stÃ¼ne deÄŸer yaz
for bar, puan in zip(bars, ortalama_puan):
    plt.text(bar.get_x() + bar.get_width()/2., bar.get_height() + 0.05,
             f'{puan}', ha='center', va='bottom', fontweight='bold')

plt.title("Departman Ortalama Performans PuanlarÄ±",
          fontsize=14, fontweight="bold")
plt.ylabel("Puan (5 Ã¼zerinden)", fontsize=12)
plt.ylim(0, 5)
plt.tight_layout()
plt.show()
```

### 7.1 YÄ±ÄŸÄ±lmÄ±ÅŸ Ã‡ubuk Grafik

YÄ±ÄŸÄ±lmÄ±ÅŸ (stacked) Ã§ubuk grafikler, birden fazla veri serisini Ã¼st Ã¼ste koyarak karÅŸÄ±laÅŸtÄ±rma imkanÄ± sunar.

```python
# HR Ã–rnek: YÄ±llara gÃ¶re iÅŸe alÄ±m / Ã§Ä±kÄ±ÅŸ
yillar = ['2021', '2022', '2023', '2024', '2025']
ise_alim = [45, 60, 55, 70, 65]
isten_cikis = [20, 25, 30, 22, 28]

plt.figure(figsize=(10, 6))
plt.bar(yillar, ise_alim, label='Ä°ÅŸe AlÄ±m', color='#2ECC71')
plt.bar(yillar, isten_cikis, bottom=ise_alim,
        label='Ä°ÅŸten Ã‡Ä±kÄ±ÅŸ', color='#E74C3C')

plt.title("YÄ±llÄ±k Ä°ÅŸe AlÄ±m ve Ã‡Ä±kÄ±ÅŸ OranlarÄ±",
          fontsize=14, fontweight="bold")
plt.ylabel("KiÅŸi SayÄ±sÄ±", fontsize=12)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```

### 7.2 Yatay Ã‡ubuk Grafik

```python
# HR Ã–rnek: En yÃ¼ksek maaÅŸlÄ± pozisyonlar
pozisyonlar = ['CTO', 'CFO', 'VP Engineering', 'KÄ±demli MÃ¼hendis',
               'Pazarlama MÃ¼dÃ¼rÃ¼', 'Ä°K MÃ¼dÃ¼rÃ¼', 'Analist']
maaslar = [55000, 48000, 42000, 35000, 30000, 28000, 22000]

plt.figure(figsize=(10, 6))
plt.barh(pozisyonlar, maaslar, color='#2E86C1', edgecolor='white')
plt.title("Pozisyon BazÄ±nda Ortalama MaaÅŸ",
          fontsize=14, fontweight="bold")
plt.xlabel("AylÄ±k MaaÅŸ (TL)", fontsize=12)
plt.tight_layout()
plt.show()
```

## 8. Alan GrafiÄŸi (Area Plot / Stack Plot)

Alan grafikleri, Ã§izgi grafiÄŸine benzer ancak altÄ± doldurulmuÅŸ olarak gÃ¶rÃ¼nÃ¼r. Zaman iÃ§indeki deÄŸiÅŸimleri ve birbirine baÄŸlÄ± gruplarÄ±n toplam katkÄ±sÄ±nÄ± gÃ¶stermek iÃ§in kullanÄ±lÄ±r. `stackplot()` fonksiyonu ile Ã§izilir.

```python
# HR Ã–rnek: Departman bazÄ±nda aylÄ±k eÄŸitim saatleri
aylar = range(1, 13)
teknik = [20, 25, 30, 22, 28, 35, 40, 38, 30, 25, 20, 22]
liderlik = [10, 12, 15, 18, 20, 22, 18, 15, 12, 10, 8, 10]
uyum = [5, 8, 10, 5, 8, 10, 5, 8, 10, 5, 8, 10]

plt.figure(figsize=(10, 6))
plt.stackplot(aylar, teknik, liderlik, uyum,
              labels=['Teknik EÄŸitim', 'Liderlik EÄŸitimi', 'Uyum EÄŸitimi'],
              colors=['#3498DB', '#E74C3C', '#2ECC71'],
              alpha=0.8)

plt.title("AylÄ±k EÄŸitim Saatleri DaÄŸÄ±lÄ±mÄ±",
          fontsize=14, fontweight="bold")
plt.xlabel("Ay", fontsize=12)
plt.ylabel("EÄŸitim Saati", fontsize=12)
plt.legend(loc='upper left', fontsize=11)
plt.xticks(range(1, 13), ['Oca','Åžub','Mar','Nis','May','Haz',
                           'Tem','AÄŸu','Eyl','Eki','Kas','Ara'])
plt.tight_layout()
plt.show()
```

## 9. Ã‡oklu Grafik ile HR Dashboard

`subplot()` fonksiyonu, tek bir gÃ¶rsel Ã¼zerinde m Ã— n dÃ¼zeninde birden fazla grafik oluÅŸturmayÄ± saÄŸlar. ÃœÃ§ parametre alÄ±r: satÄ±r sayÄ±sÄ±, sÃ¼tun sayÄ±sÄ± ve hÃ¼cre numarasÄ±.

- `tight_layout()`: Alt grafiklerin Ã¼st Ã¼ste binmesini Ã¶nler

```python
# HR Analitik Dashboard: 6 farklÄ± grafik tek gÃ¶rselde
np.random.seed(42)
fig = plt.figure(figsize=(16, 10))

# 1) Ä°ÅŸe AlÄ±m Trendi (Ã‡izgi Grafik)
ax1 = plt.subplot(2, 3, 1)
aylar_num = range(1, 13)
ise_alim_aylik = [12, 8, 15, 10, 18, 14, 20, 16, 12, 9, 11, 22]
plt.plot(aylar_num, ise_alim_aylik, 'b-o', linewidth=2)
plt.title('Ä°ÅŸe AlÄ±m Trendi', fontweight='bold')
plt.xlabel('Ay')
plt.ylabel('KiÅŸi')
plt.grid(True, alpha=0.3)

# 2) Departman DaÄŸÄ±lÄ±mÄ± (Pasta)
ax2 = plt.subplot(2, 3, 2)
plt.pie([120, 80, 35, 45, 60],
        labels=['MÃ¼h.','Paz.','Ä°K','Fin.','Op.'],
        autopct='%1.0f%%',
        colors=['#3498DB','#E74C3C','#2ECC71','#F39C12','#9B59B6'])
plt.title('Departman DaÄŸÄ±lÄ±mÄ±', fontweight='bold')

# 3) MaaÅŸ DaÄŸÄ±lÄ±mÄ± (Histogram)
ax3 = plt.subplot(2, 3, 3)
plt.hist(np.random.normal(15000, 4000, 340),
         bins=15, color='#3498DB', edgecolor='white')
plt.title('MaaÅŸ DaÄŸÄ±lÄ±mÄ±', fontweight='bold')
plt.xlabel('MaaÅŸ (TL)')

# 4) Performans PuanlarÄ± (Ã‡ubuk)
ax4 = plt.subplot(2, 3, 4)
plt.bar(['MÃ¼h.','Paz.','Ä°K','Fin.','Op.'],
        [4.2, 3.8, 4.0, 3.5, 3.9],
        color=['#2ECC71','#3498DB','#E74C3C','#F39C12','#9B59B6'])
plt.title('Performans PuanlarÄ±', fontweight='bold')
plt.ylabel('Puan')
plt.ylim(0, 5)

# 5) Deneyim-MaaÅŸ Ä°liÅŸkisi (DaÄŸÄ±lÄ±m)
ax5 = plt.subplot(2, 3, 5)
deneyim_d = np.random.uniform(1, 20, 100)
maas_d = deneyim_d * 1500 + np.random.normal(5000, 2000, 100)
plt.scatter(deneyim_d, maas_d, alpha=0.6, c='#9B59B6', s=40)
plt.title('Deneyim vs MaaÅŸ', fontweight='bold')
plt.xlabel('Deneyim (YÄ±l)')
plt.ylabel('MaaÅŸ (TL)')

# 6) EÄŸitim Saatleri (Alan Grafik)
ax6 = plt.subplot(2, 3, 6)
x = range(1, 13)
plt.stackplot(x, [20,25,30,22,28,35,40,38,30,25,20,22],
              [10,12,15,18,20,22,18,15,12,10,8,10],
              labels=['Teknik','Liderlik'],
              colors=['#3498DB','#E74C3C'], alpha=0.8)
plt.title('EÄŸitim Saatleri', fontweight='bold')
plt.legend(fontsize=8)

plt.suptitle('HR ANALÄ°TÄ°K DASHBOARD', fontsize=18,
             fontweight='bold', color='#1B4F72')
plt.tight_layout()
plt.show()
```

### 10.3 Kontur GrafiÄŸi (Contour Plot)
`contour()` fonksiyonu, Ã¼Ã§ boyutlu yÃ¼zeyleri iki boyutlu dÃ¼zlemde gÃ¶sterir. `contourf()` fonksiyonu ise sÄ±nÄ±rlarÄ±n iÃ§ini renk ile doldurur.

```python
# Kontur GrafiÄŸi Ã–rneÄŸi: Memnuniyet HaritasÄ±
# (MaaÅŸ ve Ä°ÅŸ-YaÅŸam Dengesi â†’ Memnuniyet Skoru)
np.random.seed(42)

maas_skalasi = np.linspace(8000, 40000, 50)
denge_skalasi = np.linspace(1, 10, 50)
X, Y = np.meshgrid(maas_skalasi, denge_skalasi)
Z = (X / 10000) + Y * 0.5 + np.random.normal(0, 0.3, X.shape)

plt.figure(figsize=(10, 6))
cs = plt.contourf(X, Y, Z, levels=15, cmap='RdYlGn')
plt.colorbar(cs, label='Memnuniyet Skoru')
plt.title("MaaÅŸ ve Ä°ÅŸ-YaÅŸam Dengesi â†’ Memnuniyet Skoru",
          fontsize=14, fontweight="bold")
plt.xlabel("MaaÅŸ (TL)", fontsize=12)
plt.ylabel("Ä°ÅŸ-YaÅŸam Dengesi (1-10)", fontsize=12)
plt.tight_layout()
plt.show()
```
